## outliers detection

```python
def handle_outliers_iqr(data, column):
    """Remove outliers using the IQR method"""
    Q1 = data[column].quantile(0.25)#17
    Q3 = data[column].quantile(0.75)#209
    IQR = Q3 - Q1#192
    lower = Q1 - 1.5 * IQR# 
    upper = Q3 + 1.5 * IQR#
    
    print(f"\nOutliers in {column}: {(data[column] < lower).sum() + (data[column] > upper).sum()}")
    print(f"Lower bound: {lower}, Upper bound: {upper}")
    
    # # Visualize before removal
    # plt.figure(figsize=(6,4))
    # sns.boxplot(x=data[column])
    # plt.title(f"Before removing outliers - {column}")
    # plt.show()
    
    # Remove outliers
    data = data[(data[column] >= lower) & (data[column] <= upper)]
    
    # Visualize after removal
    plt.figure(figsize=(6,4))
    sns.boxplot(x=data[column])
    plt.title(f"After removing outliers - {column}")
    plt.show()
    
    return data

# Apply to numeric columns
numeric_col=df.select_dtypes(include=['Int64','Float64']).columns
for col in numeric_col:
    df = handle_outliers_iqr(df, col)

print("\nNew Shape after outlier removal:", df.shape)
```

## feature engineering

```python
df['house_age'] = 2015 - df['yr_built']
df['is_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
df['total_sqft'] = df['sqft_above'] + df['sqft_basement']
df['price_per_sqft'] = df['price'] / df['sqft_living']
```

```python
df.dtypes
```

```python
X = df.drop(columns=['price'])#INPUT FEATURES
y = df['price']#OUTPUT VARIABLE
```

## encoding

```python
zipcode_mean = df.groupby('zipcode')['price'].mean()
df['zipcode_encoded'] = df['zipcode'].map(zipcode_mean)
df.drop('zipcode', axis=1, inplace=True)
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

# Step 2: Perform Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,       # 20% test data
    random_state=42,     # For reproducibility
    shuffle=True         # Shuffles before splitting
)

# Step 3: Display shapes to confirm split
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()


X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

## model building

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2_sc=r2_score(y_test,y_pred)
mae=mean_squared_error(y_test,y_pred)
mse=mean_absolute_error(y_test,y_pred)
print(f'r2 score:{r2_sc:.2f}')
print(f'mean squared error:{mae:.2f}')
print(f'mean absolute error:{mse:.2f}')
```
